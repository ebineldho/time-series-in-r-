## **Q4.** Data pre-processing and plotting using ggplot2

1. Code for creating a new data frame using package dplyr. This data frame is sliced and
re-arranged from the original dataset and contains only the columns and rows that are
relevant to this task.

```r
England <- fraud_percentage_change_by_area %>%
  select(-ncol(fraud_percentage_change_by_area)) %>%
  slice(0,3:50)

colnames(England) <- c("AreaCode","AreaName","NumOfOffences","OffenceRate")
England$OffenceRate <- as.numeric(England$OffenceRate)
England
```

```r
# Creating data frame for region
Region <- England %>%
  filter(AreaCode %in% c("E12000001", "E12000002", "E12000003",
"E12000004", "E12000005", "E12000006", "E12000007", "E12000008",
"E12000009")) %>%
  select(-1)
# Print the selected rows
print(Region)
```

```r
# Creating data frame for county
excluded_codes <- c("E12000001", "E12000002", "E12000003",
                    "E12000004", "E12000005", "E12000006",
                    "E12000007", "E12000008", "E12000009")
county <- England %>%
  filter(!AreaCode %in% excluded_codes) %>%
    select(-1)
county
```

2. Plot and explain: Which region(s) has the highest total count of offences as well as

```r
library(ggplot2)
library(viridis)
```

```r
# Plot for Number of Offences with viridis color scale
ggplot(Region, aes(x = reorder(AreaName, NumOfOffences), y = NumOfOffences, fill = NumOfOffences)) +
  geom_bar(stat = "identity") +

  labs(title = "Offences by Region", x = "Region", y = "No. of Offences") +
  scale_fill_viridis_c()

# Plot for Offence Rate with viridis color scale
ggplot(Region, aes(x = reorder(AreaName, OffenceRate), y = OffenceRate, fill = OffenceRate)) +
  geom_bar(stat = "identity") +

  labs(title = "Offence Rate per 1000 Population", x = "Region", y = "Offence Rate per 1000 Population") +
  scale_fill_viridis_c()
```
